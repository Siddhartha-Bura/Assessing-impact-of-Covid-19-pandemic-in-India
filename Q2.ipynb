```python
import json
import csv
import pandas as pd
import numpy as np
```

```python
# Reusing the first question's output file
f = open(r'neighbor-districts-modified.json')
```

```python
data = json.load(f)
```

```python
# Storing undirected graph as edge list format
# Assuming no repetition of edges, adding one edge only once
edge_list = []
for i in data.keys():
    for j in data[i]:
        if [j,i] not in edge_list:
            edge_list.append([i,j])
```

```python
edge_list = sorted(edge_list, key=lambda x: x[0])
```

```python
edge_list
```

```python
fp = open('edge-graph.csv', 'w+', newline ='')
with fp:    
    write = csv.writer(fp)
    write.writerows(edge_list)
```
