```python
udw_separated = []
for i in udw:
    udw_separated.append([i[:2], i[3:].lower(),i])
udw_separated
# Information stored as list of lists for convenience, to use later
```

```python
# Preprocessing the data as per our mc_dict and correction dictionaries.
# Stripping some unwanted spaces and Q codes as well
data_2 = {}
for item in list(data.keys()):
    possible_values = []
    field = item
    possible_values = data[item]
    if field not in list(correction.values()):
        field = field[:field.find('/')]
        field = field.replace('_district', '')
        field = field.replace('_', ' ')
        field = field.replace('–', ' ')
        field = field.replace('-', ' ')
    possible_values = [ i[:i.find('/')] if i not in list(correction.values()) else i for i in possible_values ]
    possible_values = [ i.replace('_district', '') if i not in list(correction.values()) else i for i in possible_values ]
    possible_values = [ i.replace('_', ' ') if i not in list(correction.values()) else i for i in possible_values]
    possible_values = [ i.replace('–', ' ') if i not in list(correction.values()) else i for i in possible_values]
    possible_values = [ i.replace('-', ' ') if i not in list(correction.values()) else i for i in possible_values]
    for k, v in mc_dict.items():
        if field == k:
            field = field.replace(k,v)
        possible_values = [i.replace(k,v) if i == k else i for i in possible_values]
    data_2[field] =  possible_values
    print(field, possible_values)
```

```python
# Assigning state codes
new_data = {}
for item in list(data_2.keys()):
    new_pv = []
    field = item
    possible_values = data_2[item]
    print(possible_values)
    for i in udw_separated:
        if field == i[1]:
            field = (i[2] + '.')[:-1]
            break
    for i in possible_values:
        for j in udw_separated:
            if i == j[1]:
                i = (j[2] + '.')[:-1]
                break
        new_pv.append(i)
    print (field,new_pv)
    new_data[field] =  new_pv
```

```python
s1 = set()
for i in new_data.keys():
    if i.find("Delhi") != -1:
        s1.add(i)
print(s1)
# Keys with Delhi
```

```python
s2 = set()
for i in new_data.values():
    for j in i:
        if j.find("Delhi") != -1:
            s2.add(j)
print(s2)
# Values with Delhi
```

```python
s3 = set()
#new_data1 = {}
for i in new_data.keys():
    if i.find('Delhi') != -1:
        s3 = s3.union(set(new_data[i]))
print(s3)
# Neighboring districts for Delhi
```

```python
# Since we have captured all neighboring districts of Delhi, we can now delete em all
for i in s2:
    del new_data[i]
```

```python
# Now deleting all parts of delhi and replacing with delhi 
for v in new_data.values():
    flag = 0
    for i in s2:
        if i in v:
            v.remove(i)
            flag = 1
    if flag == 1:
        v.append("DL_Delhi")
```

```python
# Merging all delhi parts into one single delhi
delhi_list = []
for i in s3:
    if i.find("Delhi") == -1:
        delhi_list.append(i)
new_data['DL_Delhi'] = delhi_list
```
