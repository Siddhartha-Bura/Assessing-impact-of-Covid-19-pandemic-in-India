```python
d_df = df.groupby(['District_Key'],as_index=False).agg({'2021-08-14.5':'sum','2021-08-14.6':'sum'})
```

```python
d_df
```

```python
# Function which comes in handy to divide(math divide) two columns row wise
def utility_fun(df,md,fd):    
    male_list = list(df[md])
    female_list = list(df[fd])
    ratio = []
    for i in range(0,len(male_list)):
        x = np.float64(female_list[i])/male_list[i]
        if math.isnan(x)  or math.isinf(x):
            x = 'NA' 
        ratio.append(str(x))
    return ratio
```

```python
d_df['vaccinationratio'] = utility_fun(d_df,'2021-08-14.5','2021-08-14.6')
```
