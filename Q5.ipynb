```python
import csv
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv(r'cowin_vaccine_data_districtwise.csv')
```

```python
len(df.columns)
```

```python
df_col = list(df.columns)
df_col = df_col[6:]
df_col
```

```python
# Function to translate date column into useful metric
def date_trans(i):
    if i.find('.') == -1:
        x = i.split('/')
        x.reverse()
        x = "-".join(x)
    else:
        x = i[:10]
        xi = i[10:]
        x = x.split('/')
        x.reverse()
        x = "-".join(x)
        x = x + xi
    return x
```

```python
# Renaming dates to make calculations on date easy
for i in df_col:
    df.rename(columns = {i:date_trans(i)}, inplace = True)
```

```python
print(df.columns)
```

```python
df_col = list(df.columns)
df_col = df_col[6:]
df_col
```

```python
df_col = list(set(df_col))
```

```python
# Removing dates which are greater than 14 aug 2021
for i in df_col:
    if i.find('.') == -1:
        if np.datetime64(i) > np.datetime64('2021-08-14'):
            df = df.drop([i], axis = 1)
    else:
        x = i[:10]
        xi = i[10:]
        if np.datetime64(x) > np.datetime64('2021-08-14'):
            x = x + xi
            df = df.drop([x], axis = 1)
```

```python
df.columns
```

```python
# Extracting all dates
df_col = [i[:10] for i in df_col]
df_col = list(set(df_col))
df_col.sort(key=lambda date: datetime.strptime(date, "%Y-%m-%d"))
df_col = df_col[:211]
df_col
```

```python
# First row has some unwanted info , so stripping it
df = df.iloc[1: , :]
```

```python
df
```

```python
df = df.fillna(0) # Replacing nan values to zero
```

```python
all_districts = list(df['District_Key'].unique())
len(all_districts)
```

```python
# Inserting dummy date values with zeros for calculation support, later on
prev_date_list = ['2021-01-10','2021-01-09']
for prev_date in prev_date_list:
    df.insert(loc=6, column=prev_date, value=0)
    for i in range(1,10):
        df.insert(loc=6+i, column=prev_date+"."+str(i), value=0)
```

```python
df
```

```python
# Transforming only required columns into integer type and them summing over them for same districts
```

```python
new_df_col = df_col.copy()
new_df_col.insert(0, '2021-01-10')
new_df_col.insert(0, '2021-01-09')
utility_dict = {}
convert_dict = {}
for i in new_df_col: 
    convert_dict[i+".3"] = int
    convert_dict[i+".4"] = int
    utility_dict[i+".3"] = 'sum'
    utility_dict[i+".4"] = 'sum'
```

```python
df = df.astype(convert_dict)
```

```python
d_df = df.groupby(['District_Key'],as_index=False).agg(utility_dict)
```

```python
d_df
```

```python
# Time intensive loop
# Core logic for calculation here is: no of cases in a week is = saturday of present week minus previous week's saturday
# For dates not in dataframe, we assume zeros for those
district_week_df = pd.DataFrame(columns = ['District_Key','weekid','dose1','dose2'])
for index, row in d_df.iterrows():
    temp_df = pd.DataFrame(columns = ['District_Key','weekid','dose1','dose2'])
    dist_id = row['District_Key']
    weekid = 1
    start_date = np.datetime64('2021-01-10')
    end_date = np.datetime64(df_col[-1])
    while (start_date < end_date):
        it2 = str(start_date + np.timedelta64(6,'D'))
        it1 = str(start_date - np.timedelta64(1,'D'))
        dose1 = row[it2+'.'+'3'] - row[it1+'.'+'3']
        dose2 = row[it2+'.'+'4'] - row[it1+'.'+'4']
        temp_df = temp_df.append(pd.DataFrame({"District_Key": dist_id,"weekid":weekid,"dose1":dose1,'dose2':dose2}, index=[0]))
        weekid = weekid + 1
        start_date = start_date + np.timedelta64(7,'D')
    district_week_df = district_week_df.append(temp_df)
```

```python
month_list = [['2021-01-16','2021-02-15'],['2021-02-16','2021-03-15'],['2021-03-16','2021-04-15'],
             ['2021-04-16','2021-05-15'],['2021-05-16','2021-06-15'],['2021-06-16','2021-07-15'],
             ['2021-07-16','2021-08-14'],['2021-08-15','2021-09-14']]
```

```python
# Time intensive loop
# Manually mentioned the dates in month_list that needs to be substracted to calculate peaks for months
district_month_df = pd.DataFrame(columns = ['District_Key','monthid','dose1','dose2'])
for index, row in d_df.iterrows():
    temp_df = pd.DataFrame(columns = ['District_Key','monthid','dose1','dose2'])
    dist_id = row['District_Key']
    month_id = 0
    start_date = np.datetime64(df_col[0])
    end_date = np.datetime64(df_col[-1])
    while (start_date < end_date) and (month_id<7):
        it2 = str(np.datetime64(month_list[month_id][1]))
        it1 = str(np.datetime64(month_list[month_id][0]))
        dose1 = int(row[it2+'.'+'3']) - int(row[it1+'.'+'3'])
        dose2 = int(row[it2+'.'+'4']) - int(row[it1+'.'+'4'])
        temp_df = temp_df.append(pd.DataFrame({"District_Key": dist_id,"monthid":month_id+1,"dose1":dose1,'dose2':dose2}, index=[0]))
        month_id = month_id + 1
    district_month_df = district_month_df.append(temp_df)
```

```python
# Grouping district values of same state into one by summing over them all
state_df = df.groupby(['State'],as_index=False).agg(utility_dict)
```

```python
# Calculating values for state taking week as a metric, using same formulae as stated earlier
state_week_df = pd.DataFrame(columns = ['State','weekid','dose1','dose2'])
for index, row in state_df.iterrows():
    temp_df = pd.DataFrame(columns = ['State','weekid','dose1','dose2'])
    state_id = row['State']
    weekid = 1
    start_date = np.datetime64('2021-01-10')
    end_date = np.datetime64(df_col[-1])
    while (start_date < end_date):
        it2 = str(start_date + np.timedelta64(6,'D'))
        it1 = str(start_date - np.timedelta64(1,'D'))
        dose1 = int(row[it2+'.'+'3']) - int(row[it1+'.'+'3'])
        dose2 = int(row[it2+'.'+'4']) - int(row[it1+'.'+'4'])
        temp_df = temp_df.append(pd.DataFrame({"State": state_id,"weekid":weekid,"dose1":dose1,'dose2':dose2}, index=[0]))
        weekid = weekid + 1
        start_date = start_date + np.timedelta64(7,'D')
    state_week_df = state_week_df.append(temp_df)
```

```python
# Calculating values for state taking month as a metric
state_month_df = pd.DataFrame(columns = ['State','monthid','dose1','dose2'])
for index, row in state_df.iterrows():
    temp_df = pd.DataFrame(columns = ['State','monthid','dose1','dose2'])
    state_id = row['State']
    month_id = 0
    start_date = np.datetime64(df_col[0])
    end_date = np.datetime64(df_col[-1])
    while (start_date < end_date) and (month_id<7):
        it2 = str(np.datetime64(month_list[month_id][1]))
        it1 = str(np.datetime64(month_list[month_id][0]))
        dose1 = int(row[it2+'.'+'3']) - int(row[it1+'.'+'3'])
        dose2 = int(row[it2+'.'+'4']) - int(row[it1+'.'+'4'])
        temp_df = temp_df.append(pd.DataFrame({"State": state_id,"monthid":month_id+1,"dose1":dose1,'dose2':dose2}, index=[0]))
        month_id = month_id + 1
    state_month_df = state_month_df.append(temp_df)
```

```python
# For overall of a district we take the last last value because the data is cumulative
overall_district_df = pd.DataFrame(columns=['districtid','dose1','dose2'])
for index, row in d_df.iterrows():
    overall_district_df = overall_district_df.append(pd.DataFrame({"districtid":row['District_Key'],"dose1":row['2021-08-14.3'],'dose2':row['2021-08-14.4']}, index=[0]))
```

```python
# For overall of a state we take the last last value because the data is cumulative
overall_state_df = pd.DataFrame(columns=['stateid','dose1','dose2'])
for index, row in state_df.iterrows():
    overall_state_df = overall_state_df.append(pd.DataFrame({"stateid":row['State'],"dose1":row['2021-08-14.3'],'dose2':row['2021-08-14.4']}, index=[0]))
```
