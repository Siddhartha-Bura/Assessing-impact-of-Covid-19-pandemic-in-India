```python
# Extracting only useful columns
df = df.iloc[1: , :]
df = df.fillna(0)
df = df[['State','District_Key','2021-08-07.3','2021-08-14.3']]
```

```python
df['rate'] = (df['2021-08-14.3']-df['2021-08-07.3'])/7 # Calculating rate/day for the last week
```

```python
# Changes that needs to be done in census dataframe
state_code = ''
for index, row in dfp.iterrows():
    if row['Level'] == 'STATE':
        state_code = statechanges[row['Name']]
    elif row['Name'] in changes.keys():
        dfp.at[index,'Name'] = state_code+"_"+changes[row['Name']]
    else:
        dfp.at[index,'Name'] = state_code+"_"+row['Name']
```

```python
# Extracting districts along with total population from census dataframe
dfp_D = dfp[dfp['Level'] == 'DISTRICT']
dfp_D = dfp_D.drop(['Level'], axis = 1)
dfp_D = dfp_D .sort_values(by=['Name'], ascending=True)
dfp_D
```

```python
# Extracting states along with total population from census dataframe
dfp_S = dfp[dfp['Level'] == 'STATE']
dfp_S = dfp_S.drop(['Level'], axis = 1)
dfp_S = dfp_S .sort_values(by=['Name'], ascending=True)
print (len(dfp_S))
dfp_S
```

```python
dfp_D.rename(columns = {'Name':'District_Key'}, inplace = True)
```

```python
# Because there is no Telangana at the time of census calculation, we separate those districts out and form telangana state
new_state_changes = {'AP_Adilabad':'TG_Adilabad',
                     'AP_Hyderabad':'TG_Hyderabad',
                    'AP_Karimnagar':'TG_Karimnagar',
                      'AP_Khammam':'TG_Khammam',
                 'AP_Mahabubnagar':'TG_Mahabubnagar',
                        'AP_Medak':'TG_Medak',
                     'AP_Nalgonda':'TG_Nalgonda',
                    'AP_Nizamabad':'TG_Nizamabad',
                  'AP_Ranga Reddy':'TG_Ranga Reddy',
                     'AP_Warangal':'TG_Warangal'}
```

```python
# Changing in population dataframe with above dictionary
for index, row in dfp_D.iterrows():
    if row['District_Key'] in new_state_changes.keys():
        dfp_D.at[index,'District_Key'] = new_state_changes[row['District_Key']]
```

```python
# Appending Telangana to state population dataframe
TG_df = dfp_D[ dfp_D['District_Key'].str.match(r'(^TG)')]
TGS_df = pd.DataFrame({"Name": "TELANGANA","TOT_P":TG_df.sum(axis = 0, skipna = True)['TOT_P']},index = [0])
dfp_S = dfp_S.append(TGS_df)
```

```python
# Merging dadra and nagar haveli, daman and diu into one state
dnh = dfp_S[dfp_S['Name']=='DADRA & NAGAR HAVELI']
dd = dfp_S[dfp_S['Name']=='DAMAN & DIU']
dfp_S =dfp_S.append(pd.DataFrame({"Name": "DADRA AND NAGAR HAVELI AND DAMAN AND DIU","TOT_P":float(dd['TOT_P'].to_string(index = False))+float(dnh['TOT_P'].to_string(index= False))},index = [1]))
```

```python
df['stateid'] = [i.lower() for i in df['State']]
```

```python
# Some more changes to get states
temp_df = dfp_D[dfp_D['District_Key']=='JK_Leh'] # JK_Leh in district to Ladakh as a state
dfp_S = dfp_S.append({"Name":"Ladakh","TOT_P":float(temp_df['TOT_P'].to_string(index = False)),"stateid":"ladakh"},ignore_index = True)
dfp_S['stateid'] = [i.replace('&','and').lower() for i in dfp_S['Name']] # Replacing & to and
dfp_S['stateid'] = [i.replace('NCT OF DELHI','DELHI').lower() for i in dfp_S['Name']] # To match delhi
dfp_S['Name'] = [i.replace('NCT OF DELHI','DELHI') for i in dfp_S['Name']]
```

```python
# Merging both vaccination and population dataframes
state = pd.merge(dfp_S, df, on='stateid')
state = state[['State','2021-08-07.3','2021-08-14.3','TOT_P','rate']]
state = state.astype({'2021-08-07.3':float,'2021-08-14.3':float,'TOT_P':float,'rate':float})
state.rename(columns = {'State':'stateid'}, inplace = True)
```

```python
main_df = pd.DataFrame(columns = ['stateid', 'populationleft', 'rate', 'date'])
start_date = np.datetime64('2021-08-14')
for index, row in state.iterrows():
    if row['2021-08-14.3'] < row['TOT_P']: # Checking if any population left for vaccination
        populationleft = row['TOT_P'] - row['2021-08-14.3']
        days = math.ceil(populationleft/row['rate']) # Taking ceil of number of days
        date = start_date + np.timedelta64(days,'D') # Adding no of days to the last date 14 aug 2021
        main_df = main_df.append(pd.DataFrame({"stateid":row['stateid'],"populationleft":populationleft,"rate":row['rate'],"date":date},index=[0]))
```

```python
main_df.rename(columns = {'rate':'rateofvaccination'}, inplace = True)
```

```python
main_df= main_df .sort_values(by=['stateid'], ascending=True)
```

```python
main_df.to_csv (r'complete-vaccination.csv', index = None, header=True)
```
